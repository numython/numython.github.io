```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pygmsh as pg

geom = pg.Geometry()
n = 0.2
p1 = geom.add_point((0,0,0),n)
p2 = geom.add_point((1,0,0),n)
p3 = geom.add_point((1,1,0),n)
p4 = geom.add_point((0,1,0),n)
l1 = geom.add_line(p1,p2)
l2 = geom.add_line(p2,p3)
l3 = geom.add_line(p3,p4)
l4 = geom.add_line(p4,p1)
ll = geom.add_line_loop((l1,l2,l3,l4))
geom.add_plane_surface(ll)

p,c = pg.generate_mesh(geom)
plt.triplot(p[:,0], p[:,1], c["triangle"])
```
